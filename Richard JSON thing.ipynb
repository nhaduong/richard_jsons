Convert the broken JSON files into real ones by turning the entire file into a list and making each line a list entry with a comma at the end

```python
test_file = '/Users/quakuma/Downloads/part-m-00000.json'
def reformatjson(file):
    with open(file, 'r', encoding='utf-8') as f:
        data = f.readlines()
        
    with open('/Users/quakuma/Downloads/new_richard_file.json', 'w', encoding='utf-8') as f:
        first = '[' + data[0][:-1] + ',' #get rid of newline
        f.write(first + '\n')
        for line in data[1:-1]:
            f.write(line[:-1] + ',' + '\n')
        last = data[4] + ']'
        f.write(last)
```

```python
reformatjson(test_file)
```

Flatten the loaded JSON file (as much as possible). You can see that some cells still have nested data such as entities.user_mentions

```python
jn = json_normalize(data)
```

Create a list where each entry is a csv string of the entire row

```python
los = []
for index, row in jn.iterrows():
    los.append(row.to_csv())
```

Make sure we have the same number of csv strings as rows

```python
len(los)
```

Insert the list of csv strings into the original dataframe at the end of the dataframe with column name row_string

```python
jn['row_string'] = los
```

Dump the pandas dataframe to a pickle (binary) file

```python
import pickle

with open('richard_pandas_db.pkl', 'wb') as f:
    pickle.dump(jn, f)
```
